### Ajouter une colonne "age"

Utilisez la date de naissance pour ajouter une colonne 'age'. Vous devrez peut-être calculer en année car TimeDelta est en jours par défaut.

```python
import datetime as dt

# la solution naturelle dans l'esprit de Pandas et Numpy
maires['age'] = (dt.datetime.now() - maires['naissance']).astype('timedelta64[Y]')
maires.head()
```

Cela semble bien puisque le premier maire a bien 67 ans (le 26/02/19) et non 68 si on avait fait

```python
 ((dt.datetime.now() - maires['naissance']).dt.days // 365)[0]
 
```

Il existe un paquet qui gère les dates et intervales de facon tres fine : `relativedelta` de `dateutil`. Vérifions qu'il
trouve le même résultat :

```python
from dateutil.relativedelta import relativedelta

maires['age2'] = [relativedelta(dt.datetime.now(), x).years for x in maires.naissance]
```

```python
maires[maires.age != maires.age2]
```

On a donc un problème d'un jour avec `timedelta64[Y]` ! Notons que ce problème ne semble pas lié au nombre d'années.
On prend donc la bonne version :

```python
maires['age'] = maires['age2']
maires.drop(columns='age2', inplace = True)
maires.head()
```

### Group data

Let's group all cities of the same department and

* sum the population with `np.sum`
* average the age of mayors with `np.mean`
* count the number of cities with `np.size`


### Regrouper les données

Grouper toutes les villes du même département et

* additionner la population avec `np.sum`
* calculer l'âge moyen des maires avec `np.mean`
* compter le nombre de villes avec `np.size`

Le résultat est un DataFrame avec les 3 colonnes indiquées ci-dessus.

```python
import numpy as np

maires.groupby('code dep').agg({'population':np.sum, 'age':np.mean, 'nom commune':np.size}).rename(columns={'nom commune':'nb communes'})
```
